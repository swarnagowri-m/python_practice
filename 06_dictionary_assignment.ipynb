```python
print("find the key with maximum value in a dict")
scores = {'alice': 85, 'bob': 92, 'charlie': 78, 'diana': 95, 'eve': 100}
max_key = max(scores, key=scores.get)
max_value = scores[max_key]
print(f"dictionary: {scores}")
print(f"key with maximum value: {max_key}")
print(f"maximum value: {max_value}")
```

```python
# Question 9: Count frequency of each character in string "hello"
freq = {}
string = 'hello'
for ch in string:
    if ch in freq:
        freq[ch] += 1
    else:
        freq[ch] = 1
print(freq)
```
