### tuple
- tuple are imutable (unchangeable ) , ordered sequence of items . they are defined using parentheses () , with item separated by commas. like list ,they can contains items of different data types . Because they are immutable they are often used for fixed collections of items and can be used as keys in dictionaries(wheres lists cannot)

## common Operations & Methods 
- Tuples supports indexing , slicing, len(), and membership testing(in , notin ) just like lists.However, they lack methods that modify the tuple(like append, remove,sort, etc).They have only two primary methods : count() and index()

### tuple.count(item)
- This returns the number of times the specified item appears in the tuple

```python
tech_skills = ("python","java","GIT","python","java","GIT")
```

```python
# to count numbers of python present in tuple
python_count = tech_skills.count("python")
python_count
```
