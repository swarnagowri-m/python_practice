 ## Reverse a Number (Medium)
Write a program that reverses the digits of a given number using a while loop.

Example: Input: 12345, Output: 54321

Hint: Initialize reverse=0. While n>0: get last digit (n%10), add to reverse (reverse=reverse*10+digit), remove last digit (n=n//10)

```python
# Input number
n = int(input("Enter a number: "))
reverse = 0
while n > 0:
    digit = n % 10          # Get last digit
    reverse = reverse * 10 + digit   # Add digit to reverse
    n = n // 10             # Remove last digit

print("Reversed number:", reverse)
```

## Sum of Digits 
Write a program that calculates the sum of all digits in a given number.

Example: Input: 12345, Output: 15 (1+2+3+4+5)

Hint: While n>0: add last digit to sum (sum += n%10), then remove it (n = n//10).

```python
# Input number
n = int(input("Enter a number: "))
total = 0
while n > 0:
    digit = n % 10      # Get last digit
    total += digit      # Add digit to total
    n = n // 10         # Remove last digit
print("Sum of digits:", total)
```

## Number Guessing Game (Medium)
Write a number guessing game where:

The program generates a random number between 1 and 100
The user has to guess the number
After each guess, the program tells the user if their guess is too high or too low
The game continues until the user guesses correctly
Display the number of attempts taken
Hint: Use import random and random.randint(1, 100) to generate number. Use while loop with condition guess != target. Count attempts.

```python
import random
number = random.randint(1, 100)
guess = int(input("Guess a number between 1 and 100: "))
attempts = 1
while guess != number:
    if guess < number:
        print("Too low!")
    else:
        print("Too high!")
    
    guess = int(input("Try again: "))
    attempts += 1

print("Correct! üéâ")
print("You guessed it in", attempts, "attempts.")
```

## Palindrome Checker (Medium) 
Write a program to check if a given number is a palindrome or not.

 A palindrome number reads the same forwards and backwards. Examples: 121, 12321, 1331 are palindromes Hint: Store original number, reverse it (like Problem 3.1), then compare original with reversed number.

```python
# Take input
n = int(input("Enter a number: "))

original = n
reverse = 0

# Reverse the number
while n > 0:
    digit = n % 10
    reverse = reverse * 10 + digit
    n = n // 10

# Check palindrome
if original == reverse:
    print("Palindrome number ‚úÖ")
else:
    print("Not a palindrome number ‚ùå")
```

## Armstrong Number (Hard)
Write a program to check if a number is an Armstrong number (also called Narcissistic number).

An Armstrong number is a number that is equal to the sum of its own digits each raised to the power of the number of digits.

Examples:

153 = 1¬≥ + 5¬≥ + 3¬≥ = 1 + 125 + 27 = 153 (Armstrong number)
9474 = 9‚Å¥ + 4‚Å¥ + 7‚Å¥ + 4‚Å¥ = 6561 + 256 + 2401 + 256 = 9474 (Armstrong number)
Hint: First count digits (power). Store original. Extract each digit, raise to power, add to sum. Compare sum with original.

```python
# Take input
n = int(input("Enter a number: "))

original = n
temp = n
count = 0

# Step 1: Count number of digits
while temp > 0:
    count += 1
    temp = temp // 10

# Step 2: Calculate sum of digits raised to power 'count'
temp = n
total = 0

while temp > 0:
    digit = temp % 10
    total += digit ** count
    temp = temp // 10

# Step 3: Check Armstrong
if total == original:
    print("Armstrong Number ‚úÖ")
else:
    print("Not an Armstrong Number ‚ùå")
```

## Pattern Printing - Right Triangle (Easy)
Print the following pattern for n rows:

*
**
***
****
*****
(Example shown for n=5)

Hint: Use nested loops: outer loop for rows (i from 1 to n), inner loop for columns (j from 1 to i). Print * in inner loop.

```python
# Take input
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):        # Outer loop for rows
    for j in range(1, i + 1):    # Inner loop for columns
        print("*", end="")
    print()  # Move to next line
```

## Pattern Printing - Inverted Right Triangle (Easy)
Print the following pattern for n rows:

*****
****
***
**
*
(Example shown for n=5)

Hint: Outer loop for rows. Inner loop prints stars from 1 to (n-i+1), where i is the row number.

```python
# Take input
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):              # Outer loop for rows
    for j in range(1, n - i + 2):      # Inner loop for stars
        print("*", end="")
    print()  # Move to next line
```

## Pattern Printing - Pyramid (Medium)
Print the following pyramid pattern for n rows:

    *
   ***
  *****
 *******
*********
(Example shown for n=5)

Hint: For row i: print (n-i) spaces, then print (2*i-1) stars. Use nested loops or string multiplication.

```python
# Take input
n = int(input("Enter number of rows: "))

for i in range(1, n + 1):
    # Print spaces
    for j in range(1, n - i + 1):
        print(" ", end="")
    
    # Print stars
    for k in range(1, 2 * i):
        print("*", end="")
    
    print()  # Move to next line
```

## Multiplication Table Grid (Medium)
Print a multiplication table grid from 1 to n.

Example for n=5:

1   2   3   4   5
2   4   6   8  10
3   6   9  12  15
4   8  12  16  20
5  10  15  20  25
Hint: Nested loops: for i in range(1, n+1) and for j in range(1, n+1). Print i*j with proper spacing

```python
# Take input
n = int(input("Enter value of n: "))

for i in range(1, n + 1):          # Rows
    for j in range(1, n + 1):      # Columns
        print(i * j, end="\t")     # \t gives proper spacing
    print()   # Move to next line
```

## Diamond Pattern (Hard)
Print a diamond pattern for n rows (n should be odd):

    *
   ***
  *****
 *******
*********
 *******
  *****
   ***
    *
(Example shown for n=9)

Hint: Print pyramid (upper half) for rows 1 to (n+1)//2, then print inverted pyramid (lower half) for remaining rows

```python
# Take input (n should be odd)
n = int(input("Enter an odd number of rows: "))

mid = (n + 1) // 2   # Middle row

# Upper half (Pyramid)
for i in range(1, mid + 1):
    print(" " * (mid - i) + "*" * (2 * i - 1))

# Lower half (Inverted Pyramid)
for i in range(mid - 1, 0, -1):
    print(" " * (mid - i) + "*" * (2 * i - 1))
```
