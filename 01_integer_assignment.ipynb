```python
# Question 6: Check if 97 is a prime number
print("\nQuestion 6: Check if 97 is a prime number")
num = 97
c = True
for i in range(2,num):
    if num % i == 0:
        c = False
        break
if c:
    print("prime number")
else:
    print(" not prime number")    
```
