Adds an iteams from an iterable (like another list,tuple or String) to the end of the current list.Modifies the list in-place .Differs from append which adds the itterable which adds the iterable as a single element

```python
data_science = ["python","Data intro","python for Ds"]
data_engineering = ["DE Intro","SQL for DE","ADF"]
new_tool = ("Docker", "kubernetes")
```

```python
# use append 
data_science = ["python","Data intro","python for Ds"]
data_engineering = ["DE Intro","SQL for DE","ADF"]
new_tool = ("Docker", "kubernetes")

print(data_science)
data_science.append(data_engineering)
print(data_science)
```

```python
# use append 
data_science = ["python","Data intro","python for Ds"]
data_engineering = ["DE Intro","SQL for DE","ADF"]
new_tool = ("Docker", "kubernetes")



print(data_science)
data_science.extend(data_engineering)
print(data_science)
```

## List.sort(key= none, reverse = False)
- Sorts the item of the list in-place
- Key : A function to be called on each element before making comparisons (eg key= str.lower for lower case-insentive sort)
- reverse = True : Sorts in descending order. 
- Items must be comparable (e.g all numbers or all strings)


```python
course_list = ["data science","data engineering", "Azure Devops", "Cloud Intro" ]
tools = ["python","SQL","GIT"]
student_counts = [350,420,300,500]
```

# sort course list alphabetically (in-place)
course_list = ["data science","data engineering", "Azure Devops", "Cloud Intro" ]
print(f"course list before Sorting {course_list}")
course_list.sort()
print(f"course list After Sorting {course_list}")

```python
#convert all the words into lower case and perform the actions
course_list = ["data science","data engineering","apple" ,"Azure Devops", "Cloud Intro" ]
course_list.sort(key=str.lower)
print(course_list)
```

```python
#Based on the length
course_list = ["data science","data engineering","apple","ss" ,"Azure Devops", "Cloud Intro" ]
course_list.sort(key=str.__len__)
print(course_list)
```

```python
#Based on the length
course_list = ["data science","data engineering","apple","ss" ,"Azure Devops", "Cloud Intro" ]
course_list.sort(key=len , reverse=True)
print(course_list)
```

```python
student_counts = [350,420,300,500]
print(f"student count list before sorting {student_counts}")
student_counts.sort()
print(f"student count list After sorting {student_counts}")
```

```python
student_counts = [350,420,300,500]
print(f"student count list before sorting {student_counts}")
student_counts.sort(reverse=True)
print(f"student count list After sorting {student_counts}")
```

```python
# sort list of lists based on first element
batches = [[102,'DE'],[101,'DS'],[103, 'ADD']]
print(batches)
batches.sort()
print(batches)
```

```python
batches.sort(reverse=True)
print(batches)
```

```python
#Based on the second element
batches = [[102,'DE'],[101,'DS'],[103, 'ADD'],[104,'AA']]
print(batches)
batches.sort(key=lambda x : x[1])
print(batches)
```

```python
#Based on the second element
batches = [[102,'DE','A'],[101,'DS','B'],[103, 'ADD','C'],[104,'AA','D']]
print(batches)
batches.sort(key=lambda x : x[1])
print(batches)
```

```python
#Based on the third element
batches = [[102,'DE','A'],[101,'DS','B'],[103, 'ADD','C'],[104,'AA','D']]
print(batches)
batches.sort(key=lambda x : x[2])
print(batches)
```
