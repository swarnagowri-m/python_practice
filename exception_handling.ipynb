## Exceptional Handling
- Exception handling in Python is used to handle errors that occur during program execution so that the program does not crash and can respond gracefull.

```python
print("line 1")
print("Line 2")
print(10/0)
print("line 3")
```

#### Syntax Errors
- These occur when the code does not follow Pythonâ€™s syntax rules.
- Example: Missing colon : or incorrect indentation.

#### Runtime Errors (Exceptions)
- These occur during program execution even if the syntax is correct.
- Examples include ZeroDivisionError, ValueError, TypeError, and IndexError.

#### Logical Errors
- These occur when the program runs successfully but produces incorrect output due to wrong logic in the code.

## types of Errors
#### ZeroDivisionError
- Occurs when a number is divided by zero.
- Error: You cannot divide by zero.
### ValueError
- Occurs when the correct type is expected but the value is invalid.
- Error: Cannot convert a string like "hello" into an integer.
### TypeError
- Occurs when operations are performed on incompatible data types.
- Error: Cannot add string and integer.
### IndexError
- Occurs when trying to access an index that does not exist in a list.
- Error: Index out of range.
### KeyError
- Occurs when accessing a dictionary key that does not exist.
### NameError
- Occurs when a variable is used before it is defined.
- Error: Variable x is not defined.
### IndentationError
- Occurs when indentation (spaces or tabs) is incorrect.
