```python
## calculate avg of three diff groups 

group1_sum = 12 + 34 + 67 
group1_avg = group1_sum / 3  
print(f"Group 1 average: {group1_avg}")


group2_sum = 45 + 67 + 23
group2_avg = group2_sum / 3
print(f"Group 2 average: {group2_avg}")

group3_sum = 43 + 78 + 89
group3_avg = group3_sum / 3
print(f"Group 3 average: {group3_avg}")
```

```python
# with functions 
def calculate_average(scores):
    total = sum(scores)
    avg = total / len(scores)
    return avg
print(calculate_average([1,2,3]))
```

```python
calculate_average([566,90,8766])
```

## Write a function to find the factorial of a number.
- Input: 5
- Output: 120

```python
def factorial(n):
    if n==0:
        return 1
    return n* factorial(n-1)
```

```python
def count_vowels(s):
    count = 0
    for ch in s:
        if ch.lower() in "aeiou":
            count += 1
    return count
```

```python
count_vowels("Swarma")
```

```python
# Reverse a number
def rev(n):
    reverse = 0
    while n > 0:
        digit = n % 10
        reverse = reverse * 10 + digit
        n //= 10
    return reverse
```

```python
rev(1234)
```
