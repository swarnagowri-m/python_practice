```python
# Write a function that takes a number and returns its factorial.
def factorial(num):
    if num ==0 or num ==1:
        return 1
    else:
        return num*factorial(num-1)
```

```python
print(factorial(5))
```

```python
# Write a function that checks whether a number is prime.
def is_prime(n):
    if n <= 1:
        return False

    for i in range(2, n):
        if n % i == 0:
            return False

    return True
```

```python
is_prime(3)
```

```python
is_prime(4)
```

```python
def find_largest(numbers):
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest
```

```python
#Write a function that takes a list of numbers and returns the largest number.
def largest_num(numbers):
    largest = numbers[0]
    for num in numbers:
        if num > largest:
            largest = num
    return largest
```

### Write a function that takes a list and returns the sum of all elements without using the built-in `sum() function.
### 6. Sum of List (without `sum()`)

#### Logic
- Start with `total = 0`.
- Loop through each element.
- Add each element to `total`.
- Return `total`.


```python
def sum_list(numbers):
    total = 0
    for num in numbers:
        total += num 
    return total    
```

```python
sum_list([1,2,3,4,5,6,7,8,9])
```

### Write a function that counts the number of vowels in a string.
### 7. Count Vowels

## Logic
- Create a string of vowels (`aeiouAEIOU`).
- Set `count = 0`.
- Loop through each character.
- If character is in vowels â†’ increase `count`.
- Return `count`.

```python
def count_vowels(name):
    vowels ="aeiouAEIOU"
    count = 0
    for n in name:
        if n in vowels:
            count +=1
    return count    
```

```python
count_vowels("swarna")
```

```python
# Write a function that returns the Fibonacci sequence up to `n` terms.
def fibonacci(n):
    if n <= 0:
        return []
    elif n == 1:
        return [0]
    fib = [0, 1]
    for i in range(2, n):
        fib.append(fib[i-1] + fib[i-2])
    return fib
```

```python
print(fibonacci(10))
    
```

```python
# Write a function that checks whether a string is a palindrome.
def is_palindrome(text):
    cleaned = text.lower().replace(" ", "")
    return cleaned == cleaned[::-1]
```

```python
print(is_palindrome("madam"))       
print(is_palindrome("Racecar"))      
print(is_palindrome("Hello"))       
print(is_palindrome("nurses run"))   
```

```python
# Write a function that takes two numbers and returns their greatest common divisor (GCD).
def gcd(a, b):
    while b != 0:
        a, b = b, a % b
    return abs(a)
```

```python
print(gcd(48, 18))   
print(gcd(100, 25))
```
