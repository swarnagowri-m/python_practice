```python
def merge_count(arr):
    if len(arr) <= 1:
        return arr, 0

    mid = len(arr)//2
    left, inv_left = merge_count(arr[:mid])
    right, inv_right = merge_count(arr[mid:])

    merged = []
    i = j = inv = 0

    while i < len(left) and j < len(right):
        if left[i] <= right[j]:
            merged.append(left[i])
            i += 1
        else:
            merged.append(right[j])
            inv += len(left) - i
            j += 1

    merged.extend(left[i:])
    merged.extend(right[j:])

    return merged, inv + inv_left + inv_right


T = int(input())

for _ in range(T):
    N = int(input())
    arr = list(map(int, input().split()))

    _, swaps = merge_count(arr)

    print(swaps)
```
