```python
#Create a list of squares from 1 to 10 using list comprehension
# Traditional method
list =[]
for i in range(1,11):
    
    list.append(i**2)
print(list)
```

```python
#Given nums = [2, 4, 6, 8, 10], create a new list where each number is multiplied by 5 using list comprehension.
nums = [2, 4, 6, 8, 10]
n = []
for num in nums:
    n.append(num*5)
print(n)
```

```python
n = [num*5 for num in nums ]
n
```
